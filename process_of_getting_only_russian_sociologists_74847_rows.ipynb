# Объединение двух датафреймов. 20 июля.

Загружаем необходимые библиотеки

```python
import pandas as pd
import numpy as np
```

Загружаем датафрейм, в котором находится 66473 наблюдения (из которых только 56 тысяч – статьи именно российских социологов).

```python
all_df = pd.read_excel('final_data.xlsx')
all_df.drop(labels=['Unnamed: 0'], axis=1, inplace=True)
all_df
```

Загружаю датафрейм с довыгрузкой социологов.

```python
small_df = pd.read_csv('довыгрузка_социологов.csv')
small_df.drop('Unnamed: 0', axis=1, inplace=True)
small_df
```

В маленькой части датафрейма (18 тысяч) нет дупликатов.

```python
small_df.drop_duplicates('id')
```

Загружаю список с id статей российских социологов

```python
list_of_russian_sociologists_id = pd.read_excel('Social_доп.1.xls', sheet_name='РОС. Социологи', header=None)
```

Проверяю его длину.

```python
list_of_russian_sociologists_id = np.array(list_of_russian_sociologists_id[0])
print(list_of_russian_sociologists_id)
print(len(list_of_russian_sociologists_id)) # количество статейроссиийских социологов
```

Фильтрую большой датафрейм на принадлежность к российским социологам. Удаляются социологи из Украины, Молдавии и т. д. Число совпадает с длиной массива id статей российских социологов (56777 статей)

```python
all_df['is_russian_sociologist'] = all_df['id'].apply(lambda x: x in list_of_russian_sociologists_id)
all_df['is_russian_sociologist'].value_counts() # совпадает с длиной массива - 56777
```

Выделяю в `only_russian_sociologists` статьи только российских социологов из датафрейма `all_df`

```python
only_russian_sociologists = all_df[all_df['is_russian_sociologist'] == True]
only_russian_sociologists
```

Теперь осталось только объединить эти объекты – выделенный ранее датафрейм и `small_df`, в котором содержатся 18 тысяч наблюдений, выгруженных позднее.

```python
only_russian_sociologists = pd.concat([only_russian_sociologists, small_df])
only_russian_sociologists
```

Проверяем данные на дупликаты (по колонке `id`). Дупликатов нет.

```python
only_russian_sociologists.drop_duplicates('id')
```

Скачиваем данные

```python
only_russian_sociologists.to_csv('only_russian_sociologists_74847.csv')
only_russian_sociologists.to_excel('only_russian_sociologists_74847.xlsx')
```
