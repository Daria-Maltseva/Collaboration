# Анализ данных по дополнительной выгрузке социологов

В этом документе представлена краткая сводка по дополнительной выгрузке, произведенной Ликой 14.10.2022 - файл `final_additional_papers_120639_rows` - представляющей из себя 120634 строк, каждая из которых - статья одного из авторов, присутствовавшего в нашем датафрейме `FINAL_CORRECT_SOCIOLOGISTS` и написанная в 2010 году и позднее. Здесь также представлены вычисления, обсуждавшиеся во время созвона 18.10.2022 (на котором присутствовали Дарья, Тамара, Лика, Венеамин).

```python
import pandas as pd # импортируем пандас для работы с таблицами
```

В ячейке ниже загружен наш датафрейм с довыгрузкой всех статей за 2010 год и позднее для авторов, присутствовавших в датафрейме `FINAL_CORRECT_SOCIOLOGISTS`.

```python
df = pd.read_csv('final_additional_papers_120639_rows.csv') # загружаем файл
df.drop(axis=1, labels='Unnamed: 0', inplace=True) # удаляем мусорную колонку
df
```

```python
df['rubric'].value_counts().head(15) # самые популярные рубрики статей
```

```python
df['rubric'].value_counts().tail(20) # самые непопулярные рубрики статей
```

```python
df['rubric'].value_counts().shape[0] # всего на 120 тысяч статей - 961 рубрика
```

Ниже представлены все как-то упоминающие социлогию/общество рубрики

```python
for i in df['rubric'].value_counts().reset_index()['index']:
    if 'соц' in i.lower():
        print(i) 
```

Далее на созвоне мы решили посмотреть на статьи с рубрикой социология и проверить, сколько первых авторов этих статей (колонка `author_1_id`) встречаются в нашем старом датафрейме (тоже среди первых авторов)

```python
df[df['rubric'] == 'Социология']
```

```python
authors_sociology = list(set(list(df[df['rubric'] == 'Социология']['author_1_id']))) # всего в наших новых данных
print(len(authors_sociology)) # 2296 уникальных авторов
```

Далее мы загружаем датафрейм `FINAL_CORRECT_SOCIOLOGISTS` - в переменную `df_old`.

```python
df_old = pd.read_excel('FINAL_CORRECT_SOCIOLOGISTS.xlsx')
df_old
```

```python
df_old_authors_id = list(set(list(df_old['author_1_id']))) # получаем число уникальных авторов в FINAL_CORRECT_SOCIOLOGISTS
print(len(df_old_authors_id)) # их - 15 158 человек
```

```python
len(authors_sociology) # authors_sociology - объект, в котором находятся id первых авторов статей по социологии из новой выгрузки
```

```python
authors_sociology = [i for i in authors_sociology if str(i) != 'none'] # удаляем значение none
print(len(authors_sociology)) # всего 2295 уникальных первых авторов в статьях по социологии
```

Тут происходит логическая проверка: мы пытаемся вычислить, сколько первых авторов статей по социологии из нашей дополнительной выгрузки (от 14.10.2022, в этом ноутбуке - файл `df`) есть в нашем основном массиве данных FINAL_CORRECT_SOCIOLOGISTS (в этом ноутбуке - файл `df_old`).

Выясняется, что 329 первых (из колонки `author_1_id`) авторов статей по социологии из новой выгрузки НЕ присутствуют в старом датафрейме среди первых авторов статей по социологии, а вот 1966 авторов - присутствуют.

```python
authors_not_in_df_old = [i for i in authors_sociology if int(i) not in df_old_authors_id]
print(len(authors_not_in_df_old)) # 329 авторов есть в маленьком кусочке на 5 тысяч строк со статьями по социологии
print(len(authors_sociology)  - len(authors_not_in_df_old)) # вот столько авторов
```

Тут возникают вопросы. Почему получилось так?

Проверяем, есть ли пересечения по статьям (по id статей) в старой выгрузке и в новой. Ответ - нет, нет ни одной статьи которая встречалась бы и в старом и в новом датафрейме (`df` - `additional_papers_120639_rows` и `df_old` - `FINAL_CORRECT_SOCIOLOGISTS`).

```python
# смотрим на то, сколько id'шников статей из новой выгрузки (df) есть в старой выгрузке
papers_in_old_df = list([i for i in df_old['id'] if i != 'none'])
papers_in_new_df = list(df['id'])

papers_in_old_and_in_new = [i for i in papers_in_new_df if i in papers_in_old_df]
```

```python
print(len(papers_in_old_and_in_new)) # пересечений нет!
```
